```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

def scroll_to_bottom(driver):
    old_position = driver.execute_script("return window.pageYOffset;")
    while True:

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight-document.body.scrollHeight*0.1);")
        # Wait for the page to load
        time.sleep(3)  # This delay will depend on the connection speed and server response time
        new_position = driver.execute_script("return window.pageYOffset;")
        if new_position == old_position:
            break  # Exit the loop if the page hasn't scrolled, meaning end of page
        old_position = new_position


driver = webdriver.Chrome()
driver.get('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31');

time.sleep(3)
list_items= driver.find_elements(By.CSS_SELECTOR,"li[class='ipc-metadata-list-summary-item']")
for item in list_items:
    title_ele = item.find_elements(By.CSS_SELECTOR,"h3[class='ipc-title__text ipc-title__text--reduced']")
    title = title_ele[0].text
    print(title)

seemore = driver.find_element(by=By.XPATH, value=".//span[@class='ipc-see-more__text']")
print(seemore)

btn_element = seemore.find_element(by=By.XPATH, value=".//ancestor::button")
print(btn_element)
print(btn_element.tag_name)

scroll_to_bottom(driver)
btn_element.click()
time.sleep(3)
list_items= driver.find_elements(By.CSS_SELECTOR,"li[class='ipc-metadata-list-summary-item']")

for item in list_items[-50:]:
    title_ele = item.find_elements(By.CSS_SELECTOR,"h3[class='ipc-title__text ipc-title__text--reduced']")
    title = title_ele[0].text
    print(title)

######################################2nd iteration
scroll_to_bottom(driver)
time.sleep(3)


seemore = driver.find_element(by=By.XPATH, value=".//span[@class='ipc-see-more__text']")
print(seemore)

seemore_btn = seemore.find_element(by=By.XPATH, value=".//ancestor::button")
print(seemore_btn)
print(seemore_btn.text)

scroll_to_bottom(driver)
ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
seemore_btn.click()
time.sleep(3)

list_items= driver.find_elements(By.CSS_SELECTOR,"li[class='ipc-metadata-list-summary-item']")
#metadata= driver.find_elements(By.CSS_SELECTOR,"div[class='hMHetG dli-title-metadata']")
movies =  []
for item in list_items[-100:]:
    title_ele = item.find_elements(By.CSS_SELECTOR,"h3[class='ipc-title__text ipc-title__text--reduced']")
    title = title_ele[0].text
    print(title)
```
